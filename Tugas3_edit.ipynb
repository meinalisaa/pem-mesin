## Load Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = '/content/sample_data/used_car_data.csv'
dfA = pd.read_csv(url)

dfA
```

## 1. Handle missing values (dengan drop data)

```python
dfA.info()
```

```python
dfA.isnull().sum()
```

```python
dfA.dropna(subset=["Mileage", "Seats"], axis=0, inplace=True)
```

```python
dfA.isnull().sum()
```

## 2. Mileage, Engine, Power hilangkan satuannya

```python
dfA.head()
```

```python
for i in ["Mileage","Engine","Power"]:
  dfA[i] = dfA[i].str.split(' ').str[0]

dfA
```

## 3. Buat kolom baru yang berisi merek dari mobil

```python
dfA.head()
```

```python
dfA['Brand'] = dfA['Name'].str.split(' ').str[0]
dfA
```

## 4. Encoding categorical data

```python
dfA["Fuel_Type"].value_counts()
```

```python
fuel_dict = {"Diesel":1, "Petrol":2, "CNG":3, "LPG":4}
dfA["Fuel_Type"] = dfA["Fuel_Type"].replace(fuel_dict)
dfA.head()
```

```python
dfA["Transmission"].value_counts()
```

```python
pd.get_dummies(dfA, columns=['Transmission']).head()
```

```python
dfA["Owner_Type"].value_counts()
```

```python
owner_dict = {"First":1, "Second":2, "Third":3, "Fourth & Above":4}
dfA["Owner_Type"] = dfA["Owner_Type"].replace(owner_dict)
dfA.head()
```
