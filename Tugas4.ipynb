```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/content/sample_data/pokemon-data.csv')
df1
```

```python
df1.shape
```

```python
df1.columns
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
df1.fillna(0, inplace=True)
df1.isnull().sum()
```

```python
for i in ['HP','Attack','Defense','Speed','Total']:
    df1[i] = df1[i].astype('int')
```

```python
df1.drop('Total', inplace=True, axis=1)
df1.head()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(df1['Generation'])
plt.title('Jumlah Pokemon berdasarkan Generasi')
plt.show()
```

```python
plt.hist(df1['HP'])
plt.title('Rentang Health Point Pokemon dan Frekuensinya')
plt.xlabel('Health Point')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.title("Pola Hubungan Attack dan Defense pada Pokemon")
plt.scatter(df1['Attack'],df1['Defense'])
plt.tight_layout()
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.bar(df1['Type'], df1['HP'])
plt.title('Hubungan antara Tipe dan Health Point pada Pokemon')
plt.xlabel('Tipe')
plt.ylabel('Health Point')
plt.show()
```
